```python
!pip install pandas-profiling[notebook]
!pip install -U pandas-profiling
import numpy as np
import pandas as pd
from pandas_profiling import ProfileReport
```

```python
df_pass = pd.read_csv('/content/exemplo01.csv')
df_pass.drop(columns=['etapa','frames','passe'], inplace=True)
display(df_pass)
print('\n-------------------------------------------------------------------\n')
df_pass.info()
```

```python
df_pass.groupby('equipe').sum()
```

```python
df_pass.groupby('equipe').describe()
```

```python
df_pass.groupby('equipe').describe()
```

```python
grouped_A = df_pass.groupby('equipe').get_group('A')
grouped_B = df_pass.groupby('equipe').get_group('B')
```

```python
profile_pass_A = ProfileReport(grouped_A)
profile_pass_B = ProfileReport(grouped_B)
```

```python
profile_pass_A.to_file('A_passes_html')
```

```python
profile_pass_B.to_file('B_passes_html')
```
